## 연도별 데이터의 흐름 특징이 있는가 보려 했는데, 없음
* Normalization을 위해 최근의 값을 1로 하였는데, 별도의 공통점은 없어 보임

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('data/_Main_dataset.pickle', 'rb') as f:
    dataset = pickle.load(f)
```

```python
record_list = {}
for station, item in dataset.items():
    record_data = item[2].to_numpy()
    record_list[station] = record_data
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,1,1)
for index, item in enumerate(record_list.values()):
    plt.plot(item[:,0], item[:,1]/item[-1,1])
    if index == 30:
        break
```
